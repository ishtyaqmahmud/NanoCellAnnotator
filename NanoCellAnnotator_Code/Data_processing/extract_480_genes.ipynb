```python
def get_var_names(h5ad_path):
    """
    Return list of var_names from h5ad as strings (preserve order).
    """
    adata = ad.read_h5ad(h5ad_path)
    names = [str(x).strip() for x in adata.var_names if x is not None and str(x).strip() != ""]
    return names


def first_n_unique(genelist, n=480):
    """
    Return first n unique items preserving order (uppercased).
    """
    seen = set()
    out = []
    for g in genelist:
        gu = g.upper()
        if gu not in seen:
            seen.add(gu)
            out.append(gu)
        if len(out) >= n:
            break
    return out
```

```python
# Load all var_names
all_genes = get_var_names(h5ad_path)
print(f"Total genes in var_names: {len(all_genes)}")
print("First 10 genes:", all_genes[:10])
```

```python
selected_genes = first_n_unique(all_genes, 480)

print(f"Total selected (unique): {len(selected_genes)}")
selected_genes[:20]
```

```python
txt_path = output_dir / "all_genes480_name.txt"
json_path = output_dir / "all_genes480_name.json"
csv_path= output_dir / "all_genes480_name.csv"

#Save csv
with open(csv_path, "w") as f:
    f.write("genes\n")
    for g in selected_genes:
        f.write(f"{g}\n")

print("CSV saved at:", csv_path.resolve())

# Save TXT
with open(txt_path, "w") as f:
    f.write("\n".join(selected_genes))

# Save JSON
with open(json_path, "w") as f:
    json.dump(selected_genes, f, indent=2)

print("Saved files:")
print(" -", txt_path)
print(" -", json_path)
print("_", csv_path)
```
