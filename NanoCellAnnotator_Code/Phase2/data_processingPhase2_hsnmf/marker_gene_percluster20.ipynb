```python
import scanpy as sc
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print(sc.__version__)

import pandas as pd
import json
from pathlib import Path
```

```python
cluster_gene_path= "**/Data/hSNMF_all_marker_genes_all_clusters.csv"
gene_per_cluster= pd.read_csv(cluster_gene_path)
gene_per_cluster
```

### Clean and short marker genes 

```python
#Remove rows with missing critical values
gene_per_cluster= gene_per_cluster.dropna(subset=["cluster", "gene", "p_adj", "log2FC"])
#short with each cluster
gene_per_cluster= gene_per_cluster.sort_values(
    by=["cluster", "p_adj", "log2FC"],
    ascending= [True, True, False]
)
gene_per_cluster
```

### Extract top 20 marker genes per cluster

```python
TOP_marker_gene= 20

top_20_marker_gene_per_cluster= (
    gene_per_cluster.groupby("cluster", group_keys=False).head(TOP_marker_gene)
)
top_20_marker_gene_per_cluster.head(20)
```

```python
top_20_marker_gene_per_cluster.groupby("cluster").size()
```

### Save as CSV file

```python
output_dir= Path("marker_gene_percluster_hsnmf_output")
output_dir.mkdir(exist_ok=True)

csv_output_path= output_dir/ "top_20_marker_gene_per_cluster_hsnmf.csv"

top_20_marker_gene_per_cluster.to_csv(csv_output_path,index=False)

print(f"Saved CSV to: {csv_output_path}")
```

### Save as JSON file

```python
cluster_json = {}

for cluster_id, df in top_20_marker_gene_per_cluster.groupby("cluster"):
    cluster_json[int(cluster_id)] = {
        "cluster_size": int(df["cluster_size"].iloc[0]),
        "marker_genes": df["gene"].tolist()
    }
```

```python
json_output_path= output_dir/ "top_20_marker_gene_per_cluster_hsnmf.json"

with open(json_output_path, "w") as f:
    json.dump(cluster_json, f , indent=2)

print(f" Saved JSON to: {json_output_path}")
```
